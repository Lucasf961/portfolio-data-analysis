```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from wordcloud import WordCloud
from nltk.tokenize import word_tokenize, MWETokenizer
from nltk.corpus import stopwords
import unidecode
import os
import glob

plt.rcParams['figure.figsize'] = 12,8
plt.rcParams['font.size']=30
```

```python
# Ultimo arquivo csv incluído no diretório
list_of_files = glob.glob(f'{os.getcwd()}\*csv')
latest_file = max(list_of_files, key=os.path.getctime)

df_jobs = pd.read_csv(latest_file)
```

```python
# função para retornar dataframe com a lista de palavras de cada vaga 
def filtered_df(df, description):
    for index, row in df.iterrows():
        # convertendo para letras minusculas
        lower = (row.description.lower())
        # convertendo texto para tokens com a biblioteca NLTK
        token = word_tokenize(lower)
        # removendo acentos das palavras
        accents = [unidecode.unidecode(string) for string in token]
        # Multi-word-tokenizer (e.g.: power bi)
        mwe_token = MWETokenizer([('power', 'bi'),('banco','de','dados'),('postgre','sql'),('google','planilhas'),
                                  ('visualizacao','de','dados'),('capacidade','analitica'), ('trabalho','em','equipe'),
                                  ('team','work'),('design','thinking'),('critical', 'thinking'),('data', 'visualization'),
                                  ('problem', 'solving')])
        token = mwe_token.tokenize(accents)
        # eliminando palavras duplicadas
        drop = list(set(token))
        # eliminando stop words em inglês e português ('mas', 'como', 'e', 'é'.....)
        details = [word for word in drop if word not in stopwords.words(['english', 'portuguese'])]
        # adicionando a coluna description
        row.description = details
    return df



# função para filtrar palavras chave do dataframe
def count_keywords(filtered_df, description, keywords_list):
    count_keyword = pd.DataFrame(filtered_df['description'].sum()).value_counts().rename_axis('keywords').reset_index(name ='count')
    length = len(filtered_df)
    count_keyword['percentage'] = (count_keyword['count'] / length) * 100
    count_skills = count_keyword[count_keyword.keywords.isin(keywords_list)]
    return count_skills
    
    

    # função para gerar a nuvem de palavras
def word_cloud(filtered_df, keywords_list):
    # coluna para lista
    analyst_list = df_analyst.description.tolist()
    # convertendo lista de listas em uma só lista
    flat_list = [item for sublist in analyst_list for item in sublist]
    # filtrando as palavras-chave
    skills = [item for item in flat_list if item in keywords_list]
    # convertendo lista para string(a função wordcloud espera receber uma string)
    list_skills = " ".join(skills)
    # Gerando a Word cloud
    wordcloud = WordCloud(collocations=False, background_color='white').generate(list_skills)
    # display
    plt.figure(figsize=(9,4))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
    return
```

```python
# convertendo a coluna de title para letras minusculas
df_jobs['title'] = df_jobs['title'].str.lower()

# filtrando por vagas que se encaixam em analista de dados ou data analyst
df_analyst = df_jobs[df_jobs['title'].str.contains('analista de dados|data analyst')]
```
